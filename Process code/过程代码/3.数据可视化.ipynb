```python
# 导入可视化第三方库
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'SimHei' 
import seaborn as sns
import plotly.express as px
```

```python
# 导入读取数据第三方库
import pandas as pd
import pymysql
# 连接mysql数据库
conn = pymysql.connect(host='119.91.135.214', port=3306, user='earthquake', password='123456',
                           database='earthquake', charset='utf8')
sql = "select time,level,place,lot,lat,deep,citys,provinces from china_data where time>=2023"
# 读取mysql数据库的数据为dataframe
data = pd.read_sql_query(sql, conn)
```

```python
# 导入数据处理第三方库
import datetime
# 获取年份列
# 无效的日期替换成一个合法的日期,日没有0日
data['time'] = data['time'].str.replace('-00', '-01')
# 需要先将Series类型的数据转换成字符串类型，然后再调用strptime()函数进行处理。
data['time_str'] = data['time'].apply(lambda x: str(x))
data['year'] = data['time_str'].apply(lambda x: datetime.datetime.strptime(x, '%Y-%m-%d %H:%M:%S').year)
# 数据按year列升序
# data = data.sort_values(by=['year'], ascending=True)
data
```

```python
# 历年地震数中小地震（[2,4.6]）
data['年份'] = data['year']
df_cn = data[data['provinces']!='[]']

df_xiao = df_cn.query('2<=level<=4.6')
df_xiao.groupby('年份')['time'].count().to_frame('次数').reset_index()
```

```python
# 从数据中获取每年中小地震的发生次数
counts = df_xiao.groupby('年份')['time'].count()

# 绘制柱状图
plt.rcParams['figure.dpi'] = 100
plt.bar(counts.index, counts.values)

# 设置图表标题和坐标轴标签
plt.title("近10年中小地震发生次数")
plt.xlabel("年份")
plt.ylabel("发生次数")

# 显示图表
plt.show()
```

```python
# 历年地震数中大地震（[4.7, ∞]）
data['年份'] = data['year']
df_cn = data[data['provinces']!='[]']

df_big = df_cn.query('4.7<=level<=10')
df_big.groupby('年份')['time'].count().to_frame('次数').reset_index()
```

```python
# 从数据中获取每年中小地震的发生次数
counts = df_big.groupby('年份')['time'].count()

# 绘制柱状图
plt.bar(counts.index, counts.values)

# 设置图表标题和坐标轴标签
plt.title("近10年中大地震发生次数")
plt.xlabel("年份")
plt.ylabel("发生次数")

# 显示图表
plt.show()
```

```python
# 一年全球震级级别最高的十个地方
%config InlineBackend.figure_format = 'retina'
loc = data.place[:10] #x轴坐标
m = data.level[:10]  #y轴坐标
plt.bar(range(10),m,tick_label = loc) #绘制条形图
plt.ylim((1,10))
plt.title('近一年全球震级级别最高 top10')
plt.xlabel('地震具体位置')
plt.ylabel('震级级别')
# 为每个条形图添加数据值
for x,y in enumerate(list(m)):
    plt.text(x,y+0.2,'%s' %round(y,1),ha = 'center')
plt.xticks(rotation=290) #x轴名称太长发生重叠，旋转为纵向显示
plt.tight_layout() #自动控制空白边缘，以全部显示x轴名称
plt.show()
```

```python
# 统计近一年中同一地方发生地震的次数
count = data['place'].value_counts()
plt.pie(count, labels = count.keys(), labeldistance=1.4, autopct='%2.1f%%')
plt.axis('equal')
plt.legend(loc='upper left', bbox_to_anchor=(-0.1, 1))
plt.show()
```

```python
# 统计近一年地震级别高低具体的次数
count = data['level'].value_counts()
plt.pie(count, labels = count.keys(), labeldistance=1.4, autopct='%2.1f%%')
plt.axis('equal')
plt.legend(loc='upper left', bbox_to_anchor=(-0.1, 1))
plt.show()
```

```python
# 制作地震高频地区词云分析图
from wordcloud import WordCloud
plt.rcParams['figure.dpi'] = 100
df = data['place'].value_counts().to_dict()
font_path = 'simfang.ttf'
wordcloud = WordCloud(
    background_color='#F3F3F3',
    font_path=font_path,
    width=1200,
    height=800,
    margin=2,
    max_font_size=200,
    random_state=42,
    scale=2,
    colormap='viridis_r',
)
wordcloud.fit_words(df)
plt.imshow(wordcloud, interpolation='nearest')
plt.axis('off')
plt.show()
```

```python
import folium.plugins as plugins
import folium
import requests

def wgs84tobd09(lng, lat):
    """
    将 WGS84 坐标系下的经纬度坐标转换为百度坐标系下的经纬度坐标
    """
    url = f'https://api.map.baidu.com/geoconv/v1/?coords={lng},{lat}&from=1&to=5&ak=STlOU4POGL1IomW7zo6jneGKHqlPhGF4'
    response = requests.get(url)
    result = response.json()
    if result['status'] == 0:
        return result['result'][0]['x'], result['result'][0]['y']
    else:
        return None

df = data[data['level'] >= 4.7]
df['BD_LNG'], df['BD_LAT'] = zip(*df.apply(lambda row: wgs84tobd09(row['lot'], row['lat']), axis=1))

ss1 = [[latitude, longitude] for latitude, longitude in df[['BD_LAT', 'BD_LNG']].values.tolist()]
m = folium.Map([39.904989, 116.405285],
               tiles='OpenStreetMap',
               zoom_start=6,
               attr='Map data ©2021 Baidu - GS(2021)249'
              )
groups = folium.FeatureGroup('')
for l, label in zip(ss1, data['place'].to_list()):
    groups.add_child(
        folium.CircleMarker(
            location=l,
            radius=1,
            color='red',
            fill=True,
            fill_color='red',
            fill_opacity=0.4,
            popup=folium.Popup(html=label, max_width=100),
        )
    )
m.add_child(groups)
m.add_child(folium.LatLngPopup()) 
m.save('地震分布.html')
```
