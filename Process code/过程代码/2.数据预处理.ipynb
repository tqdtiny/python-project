```python
# 数据去重(因为在两次爬取数据有积累值)
usa_data1 = usa_data.drop_duplicates(subset="time")
usa_data1
```

```python
# 添加中国地震局纬度小数点后为四位，以保证纬度的标准格式
data = china_data1['lat']
lat = []
for i in data:
    data1 = '{:.4f}'.format(i).zfill(7)
    lat.append(data1)
china_data1['lat'] = lat
china_data1
```

```python
# 建模填充缺失值
print(spark_usa2.shape)
rfDf = spark_usa2.sample(frac=1)
print(rfDf.shape)
```

```python
# 导入sklearn的随机森林算法预测缺失值
from sklearn.ensemble import RandomForestRegressor
# 使用地震深度（千米）特征值非空的样本构建训练集，地震深度（千米）特征值缺失的样本构建测试集
rfDf_train = rfDf.loc[rfDf['deep'].notnull()]
rfDf_test = rfDf.loc[rfDf['deep'].isnull()]
# 原始数据集中的无缺失数值特征
col = ['place','time','place','lot','lat']
# 划分训练集数据和标签(label)
X = rfDf_train[col]
y = rfDf_train['deep']
# 训练过程
rf = RandomForestRegressor(n_estimators=100,n_jobs=60)
rf.fit(X,y)
# 预测过程
pred = rf.predict(rfDf_test[col])
```

```python
# 补充spark_usa2缺失值
spark_usa2.loc[spark_usa2['deep'].isnull(),'deep'] = pred
print("此时的地震深度(千米)特征统计指标：\n")
print(spark_usa2['deep'].describe())
print(spark_usa2)
```

```python
# 中国地震局数据中地震所在地（省市）
# 由于在原始数据中参考位置无法便捷的解析出省和市，我这边打算用经纬度信息通过百度的API来进行获取
import requests
citys = []
provinces = []
for i, location in enumerate(china_data1[['lot', 'lat']].values):
    location = str(location[1]) + ',' + str(location[0])
    #删除lot数据前面的空格
    location1 = location.lstrip()
    url = 'https://api.map.baidu.com/reverse_geocoding/v3/?'
    params = {
        'location':location,
        'ak':'你的ak',
        'output':'json',
        }
    
    r = requests.get(url, params)
    data = r.json()['result']
    city = data['addressComponent']['city']
    province = data['addressComponent']['province']
    if len(city)==0:
        city = province
    citys.append(city)
    provinces.append(province)
    print(f'\r{i+1}',end='')
china_data1['citys'] = citys
china_data1['provinces'] = provinces
china_data1
```
